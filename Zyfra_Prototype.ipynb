### Check that recovery is calculated correctly

```python
C = gold_recovery_train['rougher.output.concentrate_au']
F = gold_recovery_train['rougher.input.feed_au']
T = gold_recovery_train['rougher.output.tail_au']

gold_recovery_train['calculated_recovery'] = (C * (F - T)) / (F * (C - T)) * 100

valid_data = gold_recovery_train.dropna(subset=['rougher.output.recovery', 'calculated_recovery'])
mae = mean_absolute_error(valid_data['rougher.output.recovery'], valid_data['calculated_recovery'])

print(f'MAE: {mae:.4f}')
```
